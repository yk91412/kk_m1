```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/colab/data/data_obj.csv', index_col = 0)
```

```python
df['year'] == 2020 # product_21, price_21, quantity_21, time_21, amount_21, time_cat21
```

```python
# 2020년도에 구매한 사람들은 21년도 관련한 컬럼에는 뜨지 않게
lis = ['product_21', 'price_21', 'quantity_21', 'time_21', 'amount_21', 'time_cat21']
for i in lis:
    df.loc[df['year'] == 2020,i]=0
```

```python
# 2021년도도 마찬가지로
lis = ['product_20', 'price_20', 'quantity_20', 'time_20', 'amount_20', 'time_cat20']
for i in lis:
    df.loc[df['year'] == 2021,i]=0
```

```python
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


fig, axes = plt.subplots(5, 4, figsize=(20, 15))

columns = list(df.select_dtypes(include='number').columns)

for i, column in enumerate(columns):
    row = i // 4
    col = i % 4
    sns.histplot(df[column], ax=axes[row, col])
    axes[row, col].set_title(column)

plt.tight_layout()
plt.show()
```

```python
# 0이 많아져서 데이터 파악이 어려우므로 20/21을 나누지 않고 한 컬럼으로 재구성
import numpy as np
df['quantity'] = np.where(df['year'] == 2020, df['quantity_20'], df['quantity_21'])
```

```python
# quantity_20과 quantity_21 컬럼은 더 이상 필요 없으므로 삭제
df.drop(['quantity_20', 'quantity_21'], axis=1, inplace=True)
```

```python
import numpy as np
df['time'] = np.where(df['year'] == 2020, df['time_20'], df['time_21'])
df['amount'] = np.where(df['year'] == 2020, df['amount_20'], df['amount_21'])
df['price'] = np.where(df['year'] == 2020, df['price_20'], df['price_21'])
df['product'] = np.where(df['year'] == 2020, df['product_20'], df['product_21'])
df['time_cat'] = np.where(df['year'] == 2020, df['time_cat20'], df['time_cat21'])
```

```python
# 필요없는 컬럼 삭제
df.drop(['time_20', 'time_21','amount_20','amount_21','price_20','price_21','product_20','product_21','time_cat20','time_cat21'], axis=1, inplace=True)
```
