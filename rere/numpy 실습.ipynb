###  아래 가이드에 따라서 1000명의 고객 구매데이터를 생성하고 아래사항을 수행하세요(관측치 10000개)

#### Q1. 고객 속성(id, 성별, 연령, 거주지역) 데이터와 거래 데이터를 통합한 데이터 프레임 df를 생성하세요.
고객별 속성 데이터
- id : 1 ~ 1000 고객 번호
- gender : 0,1 정수 난수 생성
- age : 0 ~ 80사이 정수 난수 생성
- region : 1 ~ 10 사이 정수 난수 생성

거래 데이터
- no : 1 ~ 10000번 일련번호
- day : 요일
- date : 2020-01-01 ~ 2021-12-31 사이의 날짜 1000개 생성
- product_20 : 고가제품(10), 중가제품(20), 저가제품(30)로 구분제품코드 10 ~ 30
- product_21 : 고가제품(10), 중가제품(20), 저가제품(30)로 구분제품코드 10 ~ 30
- price_20 : 1000 ~ 50000 사이 실수 난수 생성
- price_21 : 1000 ~ 50000 사이 실수 난수 생성
- quantity_20 : 1 ~ 100 사이 정수 난수 생성
- quantity_21 : 1 ~ 100 사이 정수 난수 생성
- time_20 : 01 ~ 24 사이 정수 난수 생성
- time_21 : 01 ~ 24 사이 정수 난수 생성
- amount_20 : price_20 * quantity_21
- amount_21 : price_21 * quantity_21

#### Q2. df를 수정하여 5가지 이상의 인사이트를 포함한 데이터 셋으로 변환하세요

#### Q3. df 를 탐색적 분석을 통하여 인사이트를 도출하세요.

```python
# 고객별 속성 데이터 생성
import numpy as np
import pandas as pd
import random

random.seed(1)
cusid = np.arange(1,1001)
cusno = np.arange(1,1001)
gender = np.random.randint(0,2,size = 1000)
age = np.random.randint(10,81,size = 1000)
region = np.random.randint(1,11,size=1000)


id = pd.Series(cusid)
cusno = pd.Series(cusno)
gender = pd.Series(gender)
age = pd.Series(age)
region = pd.Series(region)
df = pd.concat([id,cusno,gender,age,region],axis=1)
df.columns = ['cusid','cusno','gender','age','region']
df
```

```python
df1 = pd.DataFrame(columns=['cusno','date','day','product_20','product_21','price_20','price_21','quantity_20','quantity_21','time_20','time_21','amount_20','amount_21'])
df1
```

```python
# 거래 데이터 생성
random.seed(1)
cusno=np.random.randint(1,1001,size = 10000)
date = pd.date_range(start = '2020-01-01', end = '2021-12-31', freq = 'D') # 1000개 만들기
random_date = [random.choice(date.tolist()) for _ in range(10000)]
day = [x.day_name() for x in random_date]
product_20 = np.random.choice([10,20,30],10000)
product_21 = np.random.choice([10,20,30],10000)
price_20 = np.random.uniform(1000,50000,10000).round(2)
price_21 = np.random.uniform(1000,50000,10000).round(2)
quantity_20 = np.random.randint(1,101,size = 10000)
quantity_21 = np.random.randint(1,101,size = 10000)
time_20 = np.random.randint(1,25,10000)
time_21 = np.random.randint(1,25,10000)
amount_20 = (price_20 * quantity_20).round(2)
amount_21 = (price_21 * quantity_21).round(2)
```

```python
df1['cusno'] = cusno
df1['date'] = pd.DataFrame(random_date)
df1['day'] = pd.DataFrame(day)
df1['product_20'] = pd.DataFrame(product_20)
```

```python
df1['product_21'] = pd.DataFrame(product_21)
df1['price_20'] =pd.DataFrame(price_20)
df1['price_21'] = pd.DataFrame(price_21)
df1['quantity_20'] = pd.DataFrame(quantity_20)
df1['quantity_21'] = pd.DataFrame(quantity_21)
df1['time_20'] = pd.DataFrame(time_20)
df1['time_21'] = pd.DataFrame(time_21)
df1['amount_20'] = pd.DataFrame(amount_20)
df1['amount_21'] = pd.DataFrame(amount_21)
```

```python
merge_df['total'] = merge_df['amount_20'] + merge_df['amount_21']
```

```python
# insight 1

# 나이별 구매 금액 평균
amount_by_age =merge_df.pivot_table(values=['amount_20', 'amount_21','total'], index='age', aggfunc='mean').round(2)
amount_by_age
```

```python
# 파악 하기 어려우므로 10대,20대..로 범주 나누기
bins = [10,20,30,40,50,60,70,80,90]
labels=['teenagers','twenties','thirties','forties','fifties','sixties','seventies','eighties']
# right=False 왼쪽 경계는 포함하고 오른쪽 경계는 미포함
merge_df['age_cat'] = pd.cut(merge_df['age'],bins=bins,labels =labels,right=False)
```

```python
amount_by_age =merge_df.pivot_table(values=['amount_20', 'amount_21','total'], index='age_cat', aggfunc='mean').round()
amount_by_age=amount_by_age.rename(columns={'amount_20': '20년도 평균', 'amount_21': '21년도 평균', 'total': '전체평균'})
amount_by_age.sort_values(by=list(amount_by_age.columns))
```

```python
# 자녀들에게 돈 쓸 일이 많은 30-50대들의 평균이 높도록 / 20년도 보다 21년도가 높도록 수정
def more_amount(x):
    if x['age_cat'] in ['thirties','forties','fifties']:
        return x['amount_20'] * 1.2
    else:
        return x['amount_20'] * 0.8

def more_amount1(x):
    if x['age_cat'] in ['thirties','forties','fifties']:
        return x['amount_21'] * 1.5
    else:
        return x['amount_21']


merge_df['amount_20'] = merge_df.apply(lambda x: more_amount(x) if pd.notnull(x['amount_20']) else x['amount_20'], axis=1)
merge_df['amount_21'] = merge_df.apply(lambda x: more_amount1(x) if pd.notnull(x['amount_21']) else x['amount_21'], axis=1)
```

```python
amount_by_age =merge_df.pivot_table(values=['amount_20', 'amount_21','total'], index='age_cat', aggfunc='mean').round()
amount_by_age=amount_by_age.rename(columns={'amount_20': '20년도 평균', 'amount_21': '21년도 평균', 'total': '전체평균'})
amount_by_age.sort_values(by=list(amount_by_age.columns))
```

```python
# 20/21년도 나잇대별로 그래프
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(amount_by_age.index ,amount_by_age['20년도 평균'],marker='o',label='2020 Average')
plt.plot(amount_by_age.index,amount_by_age['21년도 평균'],marker='o',label='2021 Average')

plt.title('Average by Age Category for 2020 and 2021')
plt.xlabel('Age Category')
plt.ylabel('Average')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# 그래프 보여주기
plt.show()
```

```python
pivot_20 = merge_df.pivot_table(index=['product_20','gender'],values='age',aggfunc='mean')
pivot_21 = merge_df.pivot_table(index=['product_21','gender'],values='age',aggfunc='mean')

pivot_21.index.rename(['product_20', 'gender'], inplace=True)

# left_index=True, right_index=True
# 각 피벗 테이블의 인덱스를 기준으로 병합을 수행하도록 설정
combined_pivot = pd.merge(pivot_20, pivot_21, left_index=True, right_index=True)
combined_pivot.index.names = ['product','gender']
combined_pivot.columns=['2020 age average','2021 age average']
print(combined_pivot)
```

```python
# 고가 4-60 중가 2-30 저가 10,7-80대
# 고가와 중가는 50대쯤으로 비슷하게 중가의 연령대는 2-30대로 낮추기
def age_mean(x):
    if x['product_20'] == 10 or x['product_21'] == 10:
        return x['age'] * 1.2
    elif x['product_20'] == 20 or x['product_21'] == 20:
        return x['age'] * 0.5
    else :
        return x['age'] * 1.2
merge_df['age'] = merge_df.apply(lambda x : age_mean(x), axis=1)
```
