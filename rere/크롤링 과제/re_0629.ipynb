```python
from bs4 import BeautifulSoup
import requests
from requests.packages.urllib3.exceptions import InsecureRequestWarning
import selenium

# 경고 메시지 표시하지 않도록 설정
requests.packages.urllib3.disable_warnings(InsecureRequestWarning)

# 요청 보내기
all_list = []
for year in range(2021,2024):
    for page in range(1,7):
        url = f'https://www.aistartuptop100.co.kr/page/s2/s1.php?pg={page}&&sty={year}&ct1=1'

        response = requests.get(url, verify=False)  # SSL

        soup = BeautifulSoup(response.text,'html.parser')

        all = soup.select('.subject_wrap')

        for a in all:
            all_a = a.select('.subject')
            for i in all_a:
                print(f"{year}년도 {page} = {i.text}")
                all_list.append(i.text)
        print()
```

1. 엔체인 (NChain)
**엔체인(NChain)**은 블록체인 기술과 관련된 연구 개발을 주로 수행하는 회사입니다. 주로 비트코인과 관련된 블록체인 기술 개발에 초점을 맞추고 있으며, 분산원장 기술을 활용한 다양한 애플리케이션을 개발하는 데 관심을 가지고 있습니다.

2. 엘지전자 (LG Electronics)
**엘지전자(LG Electronics)**는 대한민국을 기반으로 하는 글로벌 전자제품 제조업체입니다. 주로 TV, 세탁기, 냉장고 등의 가전제품뿐만 아니라, 모바일 디바이스, 가정용 에어컨 및 에너지 소비 관리 시스템과 같은 스마트 기기들을 제조하고 있습니다. 또한, 엘지는 홈 엔터테인먼트 및 모바일 커뮤니케이션, 기업 솔루션과 같은 다양한 분야에서도 기술을 제공하고 있습니다.

3. 텐센트 (Tencent)
**텐센트(Tencent)**는 중국에 본사를 두고 있는 글로벌 인터넷 기술 회사로, 주로 인터넷 서비스와 인터넷 기반 상품을 제공하고 있습니다. 대표적으로는 인스턴트 메시징 서비스인 위챗(WeChat), 온라인 게임, 소셜 네트워크, 디지털 광고 등 다양한 온라인 서비스를 운영하고 있습니다. 텐센트는 또한 클라우드 컴퓨팅, 인공지능, 금융 기술 등에도 투자하고 있습니다.

4. 삼성전자 (Samsung Electronics)
**삼성전자(Samsung Electronics)**는 한국을 기반으로 하는 글로벌 전자제품 제조업체로, 전 세계적으로 휴대전화, 반도체, TV 및 가전 제품, 디스플레이 패널 등을 제조하고 있습니다. 삼성전자는 또한 반도체 제조에서 세계적인 선도적 위치를 차지하고 있으며, 스마트폰 시장에서도 중요한 시장 점유율을 보유하고 있습니다. 또한, 삼성전자는 디지털 미디어 및 소프트웨어, 인공지능 등 다양한 기술 분야에서도 활발히 연구 개발을 진행하고 있습니다.

5. 퀄컴 (Qualcomm)
**퀄컴(Qualcomm)**은 미국 캘리포니아에 본사를 두고 있는 반도체 회사로, 주로 모바일과 무선 기술 분야에서 활동하고 있습니다. 특히 퀄컴은 모바일 프로세서, 통신 칩셋, 무선 기술 표준 (예: LTE, 5G) 개발과 같은 분야에서 선도적인 역할을 하고 있습니다. 또한, 퀄컴은 IoT(Internet of Things) 기술, 자율 주행 차량 기술 등에도 투자하고 있습니다.

```python

# df_category가 카테고리를 나타내는 컬럼 이름 리스트
df_ca = ['삼성','현대','엘지']

# 서브플롯으로 그래프 배치
plt.figure(figsize=(15, 5))

for idx, category in enumerate(df_ca):
    want_group1 = df[df['applicant_sgrp'] == category]
    # 각 카테고리의 값을 합하여 리스트에 저장
    l1 = []
    a = want_group1[idx].sum()
    l1.append((idx, int(a)))

    # 합을 기준으로 내림차순 정렬
    l1.sort(key=lambda x: x[1], reverse=True)

    # 상위 5개 항목 추출
    top_categories = l1[:5]

    # 서브플롯에 그래프 추가
    plt.subplot(1, 3, idx + 1)
    labels = [x[0] for x in top_categories]
    sizes = [x[1] for x in top_categories]
    explode = [0.1] * len(top_categories)

    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=240, colors=sns.color_palette('Set3'), shadow=True, explode=explode)
    plt.axis('equal')  # 원을 원형으로 유지
    plt.title(f'Top 5 Categories for {category}')  # 각 카테고리 이름으로 제목 설정

plt.tight_layout()
plt.show()
```
