```python
from glob import glob
import pandas as pd

# csv 목록 불러오기
# 해당 패턴 만족하는 모든 파일명을 스트링으로 가져와서 리스트로 가져오기
file_names = glob("/content/drive/MyDrive/colab/datasets/소상공인시장진흥공단_상가(상권)정보_*_202403.csv")

# 비어있는 데이터프레임 생성
total = pd.DataFrame()

# 모든 csv 병합하기
for file_name in file_names:
    temp = pd.read_csv(file_name, encoding='utf-8')
    total = pd.concat([total, temp], ignore_index=True)

total.reset_index(drop=True, inplace=True)
total
```

```python
total.info()
```

```python
total.to_csv('total_info.csv',index=False, encoding='utf-8')
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')
df_상권대분류 = df.상권업종대분류명.value_counts()

df_상권대분류.plot(kind='bar')
plt.xticks(rotation=45)
plt.title('상권대분류별 건수')
plt.ylabel('count')
plt.xlabel('상권대분류')
plt.show()
```

```python
import seaborn as sns

sns.countplot(data=df, x='상권업종대분류명',palette= sns.color_palette("Set3"),order=df.상권업종대분류명.value_counts().sort_values(ascending=False).index)
plt.xticks(rotation=45)
plt.show()
```
