```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

options = Options()
options.add_argument("--headless")
options.add_experimental_option('detach', True)

driver = webdriver.Chrome(options=options)

url = 'https://www.finuts.co.kr/html/bank/product.php?se=BK03'
driver.get(url)

time.sleep(2)

# 데이터를 저장할 리스트 초기화
data = []

try:
    # 모든 버튼 요소를 수집
    elements = driver.find_elements(By.CSS_SELECTOR, '.btn_product_view')
    
    for index in range(len(elements)):
        # 각 반복에서 새로 요소 리스트를 가져옴
        elements = driver.find_elements(By.CSS_SELECTOR, '.btn_product_view')
        
        # 현재 요소를 클릭
        elements[index].click()

        # title 요소를 가져와 각 부분으로 분리
        title = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[1]/div/div[1]'))
        ).text.split("\n")  

        # title 정보들 추출
        bank = title[0] if len(title) > 0 else "N/A"
        product = title[1] if len(title) > 1 else "N/A"

        # info 요소들을 하나의 문자열로 저장
        info_dict = {}
        for i in range(1, 4):
            info = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, f'/html/body/div[1]/div[1]/ul[2]/li[{i}]'))
            ).text.split("\n")

            for j in range(0, len(info), 2):
                key = info[j]
                value = info[j + 1] if j + 1 < len(info) else "N/A"
                info_dict[key] = value

        # 각 정보 추출
        how_join = info_dict.get('가입방법', "N/A")
        join_term = info_dict.get('가입대상', "N/A")
        pre_condition = info_dict.get('우대조건', "N/A")
        
        # 리스트에 데이터 추가
        data.append({
            '은행': bank,
            '상품명': product,
            '가입방법': how_join,
            '가입대상': join_term,
            '우대조건': pre_condition
        })

        driver.back()
        time.sleep(2)

except Exception as e:
    print("Error:", e)

driver.quit()

# 리스트를 DataFrame으로 변환
df = pd.DataFrame(data)

# DataFrame 출력 또는 저장
print(df)
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

options = Options()
options.add_argument("--headless")
options.add_experimental_option('detach', True)

driver = webdriver.Chrome(options=options)

url = 'https://www.finuts.co.kr/html/bank/product.php?se=BK03'
driver.get(url)

time.sleep(2)

# 데이터를 저장할 리스트 초기화
data = []

try:
    # 총 7페이지를 반복
    for page in range(7):
        # 모든 버튼 요소를 수집
        elements = driver.find_elements(By.CSS_SELECTOR, '.btn_product_view')
        
        for index in range(len(elements)):
            # 클릭할 요소를 가져오기
            element_to_click = elements[index]
            
            # 요소가 클릭 가능할 때까지 대기
            WebDriverWait(driver, 10).until(EC.element_to_be_clickable(element_to_click))

            # 클릭할 위치로 스크롤
            driver.execute_script("arguments[0].scrollIntoView(true);", element_to_click)
            time.sleep(0.5)  # 스크롤 후 대기

            # JavaScript를 사용하여 클릭
            driver.execute_script("arguments[0].click();", element_to_click)

            # title 요소를 가져와 각 부분으로 분리
            title = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[1]/div/div[1]'))
            ).text.split("\n")  

            # title 정보들 추출
            bank = title[0] if len(title) > 0 else "N/A"
            product = title[1] if len(title) > 1 else "N/A"

            # info 요소들을 하나의 문자열로 저장
            info_dict = {}
            for i in range(1, 4):
                info = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, f'/html/body/div[1]/div[1]/ul[2]/li[{i}]'))
                ).text.split("\n")

                if i == 1:  # 가입방법
                    info_dict['가입방법'] = info[1] if len(info) > 1 else "N/A"
                elif i == 2:  # 가입대상
                    info_dict['가입대상'] = info[1] if len(info) > 1 else "N/A"
                elif i == 3:  # 우대조건
                    info_dict['우대조건'] = " ".join(info[3:]) if len(info) > 3 else "N/A"

            # 각 정보 추출
            how_join = info_dict.get('가입방법', "N/A")
            join_term = info_dict.get('가입대상', "N/A")
            pre_condition = info_dict.get('우대조건', "N/A")
            
            # 리스트에 데이터 추가
            data.append({
                '은행': bank,
                '상품명': product,
                '가입방법': how_join,
                '가입대상': join_term,
                '우대조건': pre_condition
            })

            driver.back()
            time.sleep(2)

            # 요소를 새로 가져오기
            elements = driver.find_elements(By.CSS_SELECTOR, '.btn_product_view')

        # 다음 페이지로 이동
        if page < 6:  # 마지막 페이지가 아닐 경우
            next_button = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, '//*[@id="loc-01"]/div[2]/div/li[5]/a'))
            )
            next_button.click()
            time.sleep(2)  # 다음 페이지 로드 대기

except Exception as e:
    print("Error:", e)

driver.quit()

# 리스트를 DataFrame으로 변환
df = pd.DataFrame(data)

# DataFrame 출력 또는 저장
print(df)
```
